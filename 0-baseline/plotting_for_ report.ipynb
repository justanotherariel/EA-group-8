```python
import re
import glob
import csv

data_array = []

# runs,max_gens=5,50
runs,max_gens=3,50
median = np.zeros([runs,max_gens])

# path = "baseline/fitness-2023-06-???????.csv"
path = "gen_merge_baseline.csv"
i=0
for filename in glob.glob(path):
    # print(str(filename))
    for filename in glob.glob(path):
        with open(filename, 'r') as f:
            csv_reader = csv.reader(f, delimiter='\t')
            idx = 0
            for row in csv_reader:
                row_array = [tuple(map(float, item.strip('()').split(', '))) for item in row]
                data_array.extend(row_array)
                data_out = np.asarray(data_array)
                # out= (data_out[:,1]).astype(int)
                out= (data_out).astype(int)
                median[i, idx] =  float(out)
                idx+=1
            i += 1
    #
    # with open(filename, 'r') as f:
    #     csv_reader = csv.reader(f, delimiter='\t')
    #     idx=0
    #     for row in csv_reader:
    #         pattern = r'\(\d+,\s*([-+]?\d+\.\d+)\)'
    #         string = row[1]
    #         match = re.search(pattern, string)
    #         if match:
    #             out =float(match.group(1))
    #         # print(out)
    #         median[i, idx] =  float(out)
    #         idx+=1
    #     i += 1
median_avg = np.mean(median,axis=0)
median_std = np.std(median, axis=0)
plt.errorbar(range(max_gens),median_avg,yerr=median_std, fmt='-o',label="baseline")

# print('test')
# path = "hyperparameter_tuning/pop_size/gen202206????????80.csv"
# i=0
# for filename in glob.glob(path):
#     print(str(filename))
#     with open(filename, 'r') as f:
#         csv_reader = csv.reader(f, delimiter='\t')
#         idx=0
#         for row in csv_reader:
#             median[i, idx] =  float(row[3])
#             idx+=1
#         i += 1
# median_avg = np.mean(median,axis=0)
# median_std = np.std(median, axis=0)
# plt.errorbar(range(max_gens),median_avg,yerr=median_std, fmt='-o',label="80")

# print('test')
# path = "hyperparameter_tuning/pop_size/gen202206????????160.csv"
# i=0
# for filename in glob.glob(path):
#     print(str(filename))
#     with open(filename, 'r') as f:
#         csv_reader = csv.reader(f, delimiter='\t')
#         idx=0
#         for row in csv_reader:
#             median[i, idx] =  float(row[3])
#             idx+=1
#         i += 1
# median_avg = np.mean(median,axis=0)
# median_std = np.std(median, axis=0)
# plt.errorbar(range(max_gens),median_avg,yerr=median_std, fmt='-o',label="160")


plt.legend(loc="lower right")
plt.title('Population size')
plt.xlabel("Generations")
plt.ylabel("Median fitness")
plt.show()
```
